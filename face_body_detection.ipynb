```python
import cv2 #OpenCv
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/TestImages/FullBody/images.jpg')
```

```python
cv2_imshow(image)
```

```python
tracker = cv2.CascadeClassifier('/content/ClassifyItems/haarcascade_fullbody.xml')
```

```python
imaG = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # Deixa a imagem cinza pra ter que analisar apenas um canal de cor
cv2_imshow(imaG)
```

```python
detection = tracker.detectMultiScale(imaG, scaleFactor=1.01, minSize=(50,50))
print(f"{len(detection)} pessoas encontradas")
print(detection)
```

```python
for(x,y,w,h) in detection:
  cv2.rectangle(image, (x,y), (x+w,y+h), (0,255,0), 2)
cv2_imshow(image)
```
